```python
import glob
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import numpy as np
```

```python
path = '/Users/jacquescoetzee/www/mg/courts/original text/unreported/*.rtf'
folder = glob.glob(path)

print(len(folder))
```

```python
for file in folder:
    
    # open file and create document list
    
    soup = BeautifulSoup(open(file), 'html.parser')
    document_list = str(soup).split('<hr/>')
    document_list = document_list[2:]
    
    # set df features in lists
    
    sources = []
    party_one = []
    party_two = []
    case_no = []
    judges = []
    summary = []
    date_heard = []
    judgement_heard = []
    courts = []
    
    # loop over all features in each single doc in document list
    
    for doc in document_list:

        doc = BeautifulSoup(doc, 'html.parser')

        # summaries

        for i in doc:
            summaries = doc.find('div', class_='Summary')
        if summaries is None:
            summary.append(np.nan)
        else:
            summary.append(summaries.text.replace(" \\'97 ", '. '))

        # case_no

        for i in range(0, len(doc.find_all('div', class_='Case-No'))):
            cases = doc.find_all('div', class_='Case-No')[i].text
        if cases is None:
            case_no.append(np.nan)
        else:
            case_no.append(cases)

        # sources

        for i in range(0, len(doc.find_all('div', class_='CaseName'))):
            source_names = doc.find_all('div', class_='CaseName')[i].text
        if source_names is None:
            sources.append(np.nan)
        else:
            sources.append(source_names)

        # party_one

        try:
            party_one.append(source_names.split(' v ')[0])
        except IndexError:
            party_one.append(np.nan)

        # party_two

        try:
            party_two.append(source_names.split(' v ')[1])
        except IndexError:
            party_two.append(np.nan)

        # judges

        for i in range(0, len(doc.find_all('div', class_='Judge'))):
            judge = doc.find_all('div', class_='Judge')[i].text
        if judge is None:
            judges.append(np.nan)
        else:
            judges.append(judge)

        # date_heard

        for i in range(0, len(doc.find_all('div', class_='Heard-Date'))):
            date = doc.find_all('div', class_='Heard-Date')[i].text
        if date is None:
            date_heard.append(np.nan)
        else:
            date_heard.append(date)

        # judgement_heard

        for i in range(0, len(doc.find_all('div', class_='Judgment-Date'))):
            date = doc.find_all('div', class_='Judgment-Date')[i].text
        if date is None:
            judgement_heard.append(np.nan)
        else:
            judgement_heard.append(date)

        # courts

        for i in range(0, len(doc.find_all('div', class_='Court'))):

            court = doc.find_all('div', class_='Court')[i].text
        if court is None:
            courts.append(np.nan)
        else:
            courts.append(court)
    
    # create df
    
    df = pd.DataFrame()

    df['Applicant'] = party_one
    df['Defendant'] = party_two
    df['Case No'] = case_no
    df['Judges'] = judges
    df['Summary'] = summary
    df['Date Heard'] = date_heard
    df['Date Judgement'] = judgement_heard
    df['Court'] = courts
    
    # write df to csv file
    
    df.to_csv('%s_table.csv' %file)
```
